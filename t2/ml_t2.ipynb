```
#Đồ thị y=x^2
x=[i for i in range(-10,11)]
y=[i**2 for i in x]
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```
# Đồ thị y=2x+alpha
x=[i for i in range(1,10)]
y=[2*i+6*np.random.rand() for i in x]
print(x)
print(y)
plt.plot(x,y,marker="o", color="r")
plt.show()
```

```
# vẽ hai đồ thị trên một trục toạ độ 
x = [-5,8]
y = [2,10]
plt.xlim(-10,15)
plt.ylim(-2,15)
plt.plot(x,y)
x1=[i for i in range(-3,4)]
y1=[i**2 for i in x1 ]
plt.plot(x1,y1,color='r')
plt.show() 
```

```
#3. vẽ đồ thị bar thể hiện sự phân bố dữ liệu 
names = ['weak','normal','good','excellent']
values = [25,30,23,18]
plt.figure(figsize=(6, 5))
# plt.subplot(1,2,1)
plt.bar(names, values)
plt.show()
```

```
names=['group_a', 'group_b', 'group_c']
values=[10,50,100]
plt.figure(figsize=(15,3))

plt.subplot(1,3,1)
plt.bar(names, values)
plt.subplot(1,3,2)
plt.scatter(names, values)
plt.subplot(1,3,3)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```
# Đồ thị 3D
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=plt.axes(projection="3d")

ax.set_title('Test')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

x = [3,5,9,8,2,7,5]
y = [1,3,9,9,5,7,5]
z = [6,4,3,7,8,20,10]

ax.scatter(x, y, z,color='r')
plt.xlim(0,15)
plt.ylim(0,15)
ax.set_zlim(0,25)

plt.show()
```
