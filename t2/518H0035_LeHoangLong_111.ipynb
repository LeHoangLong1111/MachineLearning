```python
#3. cho một mảng gồm n chữ cái, được sinh ngẫu nhiên
# hãy đếm frequency của các chữ cái
# vẽ đồ thị bar cho các chữ cái và tần suất tương ứng 
```

```python
import string
import random
n=np.random.randint(10,31)
print(n)
my_alphabet=[]
for i in range(n):
  my_alphabet.append(random.choice(string.ascii_lowercase))
print(my_alphabet)
# k=random.choice(string.ascii_letters)
# print(k)

result={}
for item in my_alphabet:
    if (item in result):
      result[item] += 1
    else:
      result[item] = 1
x=[]
y=[]
for key, value in result.items():
  print('%s : %d' %(key, value))
  x.append(key)
  y.append(value)
print(x)
print(y)
plt.plot(x,y, marker="o", color="b", ls="-")

```

```python
#4. Sinh mảng dữ liệu số nguyên trong khoảng [m,n] theo phân bố Gaussian 
# vẽ đồ thị histogram các giá trị này 
```

```python
import scipy.stats as gau

x= np.arange(-15,16)
print(x)
xU=x+0.5
xL=x-0.5
pro=gau.norm.cdf(xU, scale=3) - gau.norm.cdf(xL, scale=3)
pro=pro/pro.sum()
nums=np.random.choice(x,size=10000, p =pro)
plt.hist(nums, bins=len(x))




```
