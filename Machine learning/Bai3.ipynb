```python
import pandas as pd
import numpy as np 
df = pd.read_csv('abalone.data')  # read data
df.info()   # show type each colums
df
```

```python
dataset = df.values
X = dataset[:,:-1]   #fetures
Y = dataset[:,-1:]   #label
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X[:,0]= LE.fit_transform(X[:,0])  # chuyen ki tu thanh encode 
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)  # chuan hoa du lieu
```

```python
from sklearn.model_selection import train_test_split # split data thanh train , valid , test set
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3) #test size = 30% and train size = 70%
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
#X_train (10 input features, 70% of full dataset)
#X_val (10 input features, 15% of full dataset)
#X_test (10 input features, 15% of full dataset)
#Y_train (1 label, 70% of full dataset)
#Y_val (1 label, 15% of full dataset)
#Y_test (1 label, 15% of full dataset)
X_train=np.asarray(X_train).astype(np.float64)   # chuyen thanh float tranh loi 
Y_train=np.asarray(Y_train).astype(np.float64) 
X_val=np.asarray(X_train).astype(np.float64) 
Y_val=np.asarray(Y_train).astype(np.float64) 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([Dense(32, activation='relu', input_shape=(8,)),
                    Dense(32, activation='relu'),  
                    Dense(1, activation='relu'),])
# structure neural network : input layer( size =8) , hidden layer 1 ( size 32),hidden 2 ( size 32) , output 
# relu function : Rectified Linear Unit , dùng train nhanh hơn
model.compile(optimizer='sgd',  # stochastic gradient descent, với multip-class classification nên dùng 
              loss='MeanSquaredLogarithmicError',   
              metrics=['accuracy'])
         #dung binary_crossentropy thi ra loss nAN
```

```python
hist = model.fit(X_train, Y_train, epochs=100,          validation_data=(X_val, Y_val))
```
